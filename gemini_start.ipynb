```python
def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

- 채팅 대화

```python
model = genai.GenerativeModel('gemini-pro')
chat = model.start_chat(history=[])
chat
```

```python
response = chat.send_message("어린이에게 컴퓨터가 어떻게 작동하는지 한 줄로 설명한다면 어떻게 할 수 있어?")
to_markdown(response.text)
```

```python
response = chat.send_message("고등학생에게 좀 더 자세하게 설명한다면 어떻게 할 수 있을까?", stream=True)

for chunk in response:
  print(chunk.text)
  print("_"*80)
```

```python
for message in chat.history:
  display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
```

- 임베딩


Task Type | Description
---       | ---
RETRIEVAL_QUERY	| Specifies the given text is a query in a search/retrieval setting.
RETRIEVAL_DOCUMENT | Specifies the given text is a document in a search/retrieval setting. Using this task type requires a `title`.
SEMANTIC_SIMILARITY	| Specifies the given text will be used for Semantic Textual Similarity (STS).
CLASSIFICATION	| Specifies that the embeddings will be used for classification.
CLUSTERING	| Specifies that the embeddings will be used for clustering.


```python
content = "데이터 과학이 뭐야?"

result = genai.embed_content(
    model="models/embedding-001",
    content=content,
    task_type="retrieval_document",
    title="Embedding of single string")

# 1 input > 1 vector output
print(str(result['embedding'])[:50], '... TRIMMED]')
```

```python
result = genai.embed_content(
    model="models/embedding-001",
    content=[
      '데이터 과학이 뭐야?',
      '컴퓨터는 어떻게 동작해?',
      '임베딩이 뭐야?'],
    task_type="retrieval_document",
    title="Embedding of list of strings")

# A list of inputs > A list of vectors output
for v in result['embedding']:
  print(str(v)[:50], '... TRIMMED ...')
```

- Encode messages

```python
import google.ai.generativelanguage as glm
```

```python
model = genai.GenerativeModel('gemini-pro-vision')
response = model.generate_content(
    glm.Content(
        parts = [
            glm.Part(text="이 사진에 대해서 블로그에서 쓸 글을 간단하게 한글로 작성해줘."),
            glm.Part(
                inline_data=glm.Blob(
                    mime_type='image/jpeg',
                    data=pathlib.Path('./imgs/food1.jpg').read_bytes()
                )
            ),
        ],
    ),
    stream=True)
```

```python
response.resolve()

to_markdown(response.text[:100] + "... [TRIMMED] ...")
```

- 연속 채팅

```python
model = genai.GenerativeModel('gemini-pro')

messages = [
    {'role':'user',
     'parts': ["어린이에게 컴퓨터가 어떻게 작동하는지 간단하게 설명한다면 어떻게 할 수 있어?"]}
]
response = model.generate_content(messages)

to_markdown(response.text)
```

```python
messages.append({'role':'model',
                 'parts':[response.text]})

messages.append({'role':'user',
                 'parts':["고등학생에게 좀 더 자세하게 설명한다면 어떻게 할 수 있을까?"]})

response = model.generate_content(messages)

to_markdown(response.text)
```

- generation configuration

```python
response = model.generate_content(
    '재미있는 이야기 하나 들려줘',
    generation_config=genai.types.GenerationConfig(
        # Only one candidate for now.
        candidate_count=1,
        stop_sequences=['x'],
        max_output_tokens=20,
        temperature=1.0)
)

response.text
```
